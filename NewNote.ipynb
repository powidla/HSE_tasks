```python
import pandas as pd

# Assuming you have a pandas DataFrame called 'train_data' with the provided structure

# Handling missing values
train_data = train_data.dropna()  # Remove rows with missing values

# # Handling categorical data
# train_data = pd.get_dummies(train_data, columns=['location_id', 'product_id'])  # One-hot encoding for location_id and product_id columns

# Scaling numerical features
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
numerical_columns = ['PRICE_REGULAR', 'PRICE_AFTER_DISC', 'NUM_CONSULTANT', 'location_id', 'product_id']
train_data[numerical_columns] = scaler.fit_transform(train_data[numerical_columns])

# Handling boolean flags
train_data['PROMO1_FLAG'] = train_data['PROMO1_FLAG'].astype(bool)  # Convert to boolean
train_data['PROMO2_FLAG'] = train_data['PROMO2_FLAG'].astype(bool)  # Convert to boolean

# Dropping irrelevant columns
train_data = train_data.drop(['period_dt', 'id'], axis=1)  # Remove 'period_dt' and 'id' columns

```
