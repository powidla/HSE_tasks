## 3. Преобразования dataframe для дальнейшего анализа данных

```python
list_ind=list(ind.iloc[[35,44,50,59,65,75,78],0]) # создаем список номеров индикаторов для дальнейшего анализа
```

```python
list_ind
```

```python
imf_reduced=imf[imf['WEO Subject Code'].isin(list_ind)].drop(imf.columns[[0,3,4,5,6,7,8]],axis=1)
# фильтр в соответствии с ранее составленным списком индикаторов, оставаляем только нужные столбцы
```

```python
imf_reduced.head()
```

```python
col_=[str(i) for i in range(1980,2018)] # вспомогательный список для  преобразования dataframe в длинный формат
```

```python
col_[:10]
```

```python
imf_long=pd.melt(imf_reduced,id_vars=['ISO','WEO Subject Code'],value_vars=col_)
# преобразуем dataframe из широкого формата в длинный
```

```python
imf_long.head(10)
```

```python
imf_long.dtypes # типы переменных в столбцах - два столбца с годом и значением индикатора имеют тип "object",
# т.е. не воспринимаются, как числовые
```

```python
imf_long['value']=imf_long['value'].apply(lambda x:str(x).replace(',',''))
# заменим запятую на пробел для дальнейшего преобразования в числовой формат
```

```python
imf_long['value']=pd.to_numeric(imf_long['value'],errors='coerce')
# преобразуем столбец со значением индиктаора в число с плавающей точкой
```

```python
imf_long['variable']=pd.to_numeric(imf_long['variable'],errors='coerce')
# преобразуем столбец со значением года в целое число 
```

```python
imf_wide=imf_long.pivot_table(index=['ISO','variable'],columns=['WEO Subject Code'],values=['value'])
# преобразуем dataframe из длинного формата в широкий, сделав выбранные индикаторы столбцами
```

```python
imf_wide.head()
```

```python
imf_wide=imf_wide.reset_index()# переиндексация dataframe
```

```python
imf_wide.head()
```

```python
imf_wide.columns=imf_wide.columns.droplevel(0)# удаление уровня в названии столбцов
```

```python
imf_wide.head()
```

```python
imf_wide.columns
```

```python
col_names=['Country','Year']+list(imf_wide.columns[2:]) 
```

```python
col_names
```

```python
imf_wide.columns=col_names # переименование столбцов
```

```python
imf_wide.head()
```

## 4. Получение различных моделей наблюдений

```python
imf_wide.Country.unique() # список уникальных кодов стран
```

```python
CS=imf_wide[(imf_wide.Country.isin(['JPN','CHN','USA','GBR','FRA']))&(imf_wide.Year==2016)].drop(['Year'],axis=1)
#  манипуляции dataframe для получения пространственной выборки
```

```python
CS.head()
```

```python
CHN_series=imf_wide[imf_wide.Country=='CHN'][['Year','NGDP_RPCH']]
#  манипуляции dataframe для прлучения временного ряда
```

```python
CHN_series.tail(10)
```

```python
Panel=imf_wide[(imf_wide.Country.isin(['JPN','CHN','USA','GBR','FRA']))&((imf_wide.Year==2010)|(imf_wide.Year==2016))]
#  манипуляции dataframe для получения панельных данных
```
